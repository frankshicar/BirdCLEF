```python
import sys
import os

# ★ UPDATE THIS PATH
MODEL_DIR         = "/kaggle/input/datasets/qqqdddqqq/birdclef-2026-trained-model-v9"
COMPETITION_DIR   = "/kaggle/input/competitions/birdclef-2026"
SOUNDSCAPE_DIR    = os.path.join(COMPETITION_DIR, "test_soundscapes")
SAMPLE_SUBMISSION = os.path.join(COMPETITION_DIR, "sample_submission.csv")
CHECKPOINT_PATH   = os.path.join(MODEL_DIR, "checkpoint.pt")
OUTPUT_PATH       = "/kaggle/working/submission.csv"
BATCH_SIZE        = 32
DEVICE            = "cpu"
TTA               = False

sys.path = [p for p in sys.path if "birdclef" not in p.lower()]
sys.path.insert(0, MODEL_DIR)
for mod in list(sys.modules.keys()):
    if "birdclef" in mod:
        del sys.modules[mod]

assert os.path.exists(CHECKPOINT_PATH), f"Checkpoint not found: {CHECKPOINT_PATH}"
assert os.path.exists(SOUNDSCAPE_DIR),  f"Soundscape dir not found: {SOUNDSCAPE_DIR}"
print("Paths OK")
```

```python
from birdclef2026.src.inference import InferenceEngine

engine = InferenceEngine(
    checkpoint_path=CHECKPOINT_PATH,
    device=DEVICE,
    batch_size=BATCH_SIZE,
    tta=TTA,
)
engine.verify_paths()
print("Engine ready")
```

```python
engine.run(
    soundscape_dir=SOUNDSCAPE_DIR,
    sample_submission_path=SAMPLE_SUBMISSION,
    output_path=OUTPUT_PATH,
)
print(f"Done: {OUTPUT_PATH}")
```
